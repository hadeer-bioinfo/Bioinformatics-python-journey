```python
dna = input("Enter DNA sequence")
rna = ""
for base in dna:
  if base == "A":
    rna += "U"
  elif base == "T":
    rna += "A"
  elif base == "G":
    rna += "C"
  elif base == "C":
    rna += "G"


print("Original DNA sequence:")
print(dna)
print("Transcribed RNA sequence:")
print(rna)
```
